```python
import pandas as pd
```

```python
def make_file(i, j):
    df = pd.read_fwf('g2-'+str(i)+'-'+str(j)+'.txt', names = list(range(i)))
    df.to_csv('g2-'+str(i)+'-'+str(j)+'.csv')
```

```python
for i in range(10):
    for j in range(10):
        make_file(pow(2,i), 10*(j+1))
```

```python
import json
def make_json(i, j):
    config = {
        'path': '../data/benchmark_data/g2-'+str(i)+'-'+str(j)+'.csv',
        'numeric_cols': list(range(i))
    }
    
    string = json.dumps(config)
    print(string,  file=open('g2-'+str(i)+'-'+str(j)+'.json', 'w'))
```

```python
#import json
for i in range(10):
    for j in range(10):
        make_json(pow(2,i), 10*(j+1))
```
